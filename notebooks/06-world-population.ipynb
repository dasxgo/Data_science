```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
world_population = pd.read_csv("/content/world_population.csv", sep=",",header=0)
world_population.head()
```

## Data


```python
df_continent = world_population.groupby('Continent').count()
df_continent['Country/Territory']
```

##  Countries with the largest population

```python
p2022 = world_population.sort_values("2022 Population",ascending=False)
p2022.head(10)
```

```python
f, ax = plt.subplots(figsize=(8, 5))
sns.set_color_codes("pastel")
sns.barplot(x="2022 Population", y="Country/Territory", data=p2022.head(10),
            label="Total", color="b")
sns.set_color_codes("muted")

ax.legend(ncol=10, loc="lower right", frameon=True)
ax.set(xlim=(0, 1000000000), ylabel="",
       xlabel="2022 Population")
sns.despine(left=True, bottom=True)
```

##   Population by continent




```python
continent = world_population[world_population['Continent'] == 'South America']
continent = continent.loc[:, ["Country/Territory", 'Continent','2022 Population']].sort_values("2022 Population",ascending=False)
continent.head()
```

```python
sns.barplot(data=continent.head(5), x='Country/Territory', y="2022 Population", 
            n_boot=1000, saturation=0.3, orient='v', seed=1, edgecolor=".5", errcolor="0.26", errwidth=0.5, capsize=0.5, 
            ci='deprecated', axes=None, dodge=True)
```
